```python
import re

def process_text(text):
    # 1. 移除所有的 \n 符號，將其替換為空格，確保句子不會因為換行被打斷
    cleaned_text = text.replace('\n', ' ')
    
    # 2. 將 \' 替換為正常的單引號 '
    cleaned_text = cleaned_text.replace("\\'", "'")
    
    # 3. 使用正則表達式根據句號進行分割，句號後面要有空格並且下一句以大寫字母開頭
    sentences = re.split(r'(?<=\.)\s+(?=[A-Z])', cleaned_text)
    
    # 4. 去掉每個句子首尾多餘的空格
    return [sentence.strip() for sentence in sentences if sentence]

# 讀取文件內容
with open('./fairy_tales/70.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 處理文本
processed_sentences = process_text(text)


```

using ner to label

```python
from gliner import GLiNER
import json

# Initialize GLiNER with the base model
model = GLiNER.from_pretrained("urchade/gliner_mediumv2.1")



# Labels for entity prediction
# Most GLiNER models should work best when entity types are in lower case or title case
labels = ["Male", "Female"]
# 輸出處理過的文本
entities = []
for sentence in processed_sentences:
    # Perform entity prediction
    entities.append(model.predict_entities(sentence, labels, threshold=0.5))


# Display predicted entities and their labels
# for entity in entities:
#     print(entity)
#     print(entity["text"], "=>", entity["label"])
    
# write in json     
# entities_list = []
# for entity in entities:
#     entities_list.append({"text": entity["text"], "label": entity["label"]})
#     print(entity["text"], "=>", entity["label"])
# with open("predicted_entities.json", "w", encoding="utf-8") as f:
#     json.dump(entities_list, f, ensure_ascii=False, indent=4)
```
